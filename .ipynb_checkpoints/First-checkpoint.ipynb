```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
events_data = pd.read_csv("event_data_train.csv")
submissions_data = pd.read_csv("submissions_data_train.csv")
submissions_data.head()
```

```python
#Нахождение степа с максимальным количеством неправильных решений
submission_wrong = submissions_data[submissions_data['submission_status']=='wrong']
submission_wrong_max = submission_wrong.groupby(['step_id', 'user_id'], as_index=False).agg({'timestamp':'max'})
submission_wrong_max.groupby('step_id', as_index=False).agg({'user_id':'count'})\
.sort_values(by = ['user_id'], ascending =  False).head()
```

```python
events_data['date'] = pd.to_datetime(events_data['timestamp'], unit = 's')
events_data['day'] = events_data['date'].dt.date
events_data.head()
```

```python
%matplotlib inline
events_data.groupby('day')['user_id'].nunique().plot()
```

```python
users_events_data = events_data.\
pivot_table(index = 'user_id', columns = 'action', values= 'step_id', aggfunc = 'count', fill_value=0)\
.reset_index()
```

```python
events_data.pivot_table(index = 'user_id', columns = 'action', values= 'step_id', aggfunc = 'count', fill_value=0)\
.reset_index()['passed'].hist()
```

```python
submissions_data['date'] = pd.to_datetime(submissions_data['timestamp'], unit = 's')
submissions_data['day'] = submissions_data['date'].dt.date
submissions_data.head()
```

```python
users_scores = submissions_data.pivot_table(index = 'user_id', columns = 'submission_status', \
                                            values= 'step_id', aggfunc = 'count', fill_value=0)\
                                            .reset_index()
users_scores.head()
```

```python
gap_data = events_data[['user_id', 'day', 'timestamp']].drop_duplicates(subset=['user_id', 'day']).\
groupby('user_id')['timestamp'].apply(list).apply(np.diff).values
gap_data = pd.Series(np.concatenate(gap_data, axis = 0))
```

```python
gap_data = gap_data/(24*60*60)
gap_data[gap_data<200].hist()
```

```python
users_data = events_data.groupby('user_id', as_index=False).\
agg({'timestamp':'max'}).rename(columns={'timestamp':'last_timestamp'})
```

```python
now = 1526772811
drop_out_threshold = 30*24*60*60
users_data['is_gone_user'] = (now - users_data['last_timestamp'])>drop_out_threshold
```

```python
users_data=users_data.merge(users_scores, on='user_id', how='outer')
```

```python
users_data=users_data.fillna(0)
users_data.head()
```

```python
users_data=users_data.merge(users_events_data, how='outer')
users_data.head()
```

```python
users_days = events_data.groupby('user_id')['day'].nunique().to_frame().reset_index()
users_data=users_data.merge(users_days, how='outer')
users_data.head()
```

```python
users_data['passed_course'] = users_data['passed']>170
users_data.head()
```

```python
users_data.groupby('passed_course').count()
```

```python
users_data[users_data['passed_course']==True]['day'].median()
```

```python
user_min_time = events_data.groupby('user_id', as_index=False).\
agg({'timestamp':'min'}).rename(columns={'timestamp':'min_timestamp'})
```

```python
users_data=users_data.merge(user_min_time, how='outer')
users_data.head()
```
